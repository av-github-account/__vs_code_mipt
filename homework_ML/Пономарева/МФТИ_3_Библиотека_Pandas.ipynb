```python
user_table = pd.read_csv('user_table.csv')
user_table.head()
```

```python
user_table.shape
```

```python
home_page = pd.read_csv('home_page_table.csv')
home_page.head()
```

```python
home_page.shape
```

```python
sum(home_page.duplicated())
```

```python
search_page = pd.read_csv('search_page_table.csv')
search_page.head()
```

```python
search_page.shape
```

```python
payment_page = pd.read_csv('payment_page_table.csv')
payment_page.head()
```

```python
payment_page.shape
```

```python
payment_confirmation = pd.read_csv('payment_confirmation_table.csv')
payment_confirmation.head()
```

```python
payment_confirmation.shape
```

```python
home_page = home_page.rename(
    columns={'page': 'home_page'}
)

search_page = search_page.rename(
    columns={'page': 'search_page'}
)

payment_page = payment_page.rename(
    columns={'page': 'payment_page'}
)

payment_confirmation = payment_confirmation.rename(
    columns={'page': 'payment_confirmation'}
)
```

```python
# Склеиваем последовательно все датасеты с помощью метода merge по идентификатору user_id
df = user_table.merge(home_page, on='user_id', how='left') \
               .merge(search_page, on='user_id', how='left') \
               .merge(payment_page, on='user_id', how='left') \
               .merge(payment_confirmation, on='user_id', how='left')
df.head()
```

```python
df.shape
```

```python
from functools import reduce
```

```python
# Второй способ: склеиваем все датасеты сразу

data_frames = [user_table, home_page, search_page, payment_page, payment_confirmation]

df_merged = reduce(
    lambda left, right: pd.merge(left, right, on='user_id', how='left'),
    data_frames
)

df_merged.head()
```

```python
df_merged.shape
```

```python
df['home_page'] = df['home_page'].apply(
    lambda x: 1 if x == 'home_page' else 0
)

df['search_page'] = df['search_page'].apply(
    lambda x: 1 if x == 'search_page' else 0
)

df['payment_page'] = df['payment_page'].apply(
    lambda x: 1 if x == 'payment_page' else 0
)

df['payment_confirmation'] = df['payment_confirmation'].apply(
    lambda x: 1 if x == 'payment_confirmation_page' else 0
)

df.head()
```

```python
df['device'].value_counts()
```

```python
df['payment_confirmation'].agg({'users': 'count', 'payers': 'sum'})
```

```python
summary = df[['home_page', 'search_page', 'payment_page', 'payment_confirmation']].sum()
summary.head()
```

```python
summary = summary.reset_index()
summary.head()
```

```python
summary = summary.rename(columns={'index': 'stage', 0: 'number'})
summary.head()
```

```python
# Импортируем графическую библиотеку Plotly (https://plotly.com/python/funnel-charts/)
from plotly import graph_objects as go

fig = go.Figure(go.Funnel(
    y=summary['stage'],
    x=summary['number'],
    textinfo="value+percent previous"
    )
)

fig.update_layout(
    title='Воронка конверсии Интернет-магазина',
    autosize=False,
    width=1000,
    height=600
)

fig.show()
```

```python
dfg = df.groupby('device')[
    ['home_page', 'search_page', 'payment_page', 'payment_confirmation']
].sum()

dfg
```

```python
dfg = dfg.unstack('device')

dfg
```

```python
dfg = dfg.unstack('device').reset_index()

dfg
```

```python
dfg.columns.name = None

dfg = dfg.rename(columns={'index': 'Page'})

dfg.head()
```

```python
fig = go.Figure()

fig.add_trace(go.Funnel(
    y=dfg['Page'],
    x=dfg['Desktop'],
    name='Desktop',
    textposition='inside',
    textinfo='value+percent previous'
    )
)

fig.add_trace(go.Funnel(
    y=dfg['Page'],
    x=dfg['Mobile'],
    name='Mobile',
    textposition='auto',
    textinfo='value+percent previous'
   )
)

fig.update_layout(
    title='Воронка конверсии по типам устройств',
    autosize=False,
    width=1000,
    height=600
)

fig.show()
```

### 🧠 Упражнения:

#### 1. Аудитория какого пола и на каких типах устройств больше всего у интернет-магазина?

```python
#@title pivot_table (ответ)
pd.pivot_table(df, index='sex', columns='device', values='home_page', aggfunc='sum', margins=True)
```

```python
#@title crosstab (ответ)
# по умолчанию рассчитывает частоты, поэтому values=df['user_id'] и aggfunc='count' можно не указывать
pd.crosstab(index=df['sex'], columns=df['device'], margins=True)
```

```python
#@title crosstab c нормировкой по столбцам (ответ)
pd.crosstab(index=df['sex'], columns=df['device'], margins=True, normalize='columns')
```

```python
#@title crosstab c нормировкой по строкам (ответ)
pd.crosstab(index=df['sex'], columns=df['device'], margins=True, normalize='index')
```

```python
#@title groupby (ответ)
df.groupby(['sex', 'device'])['user_id'].count().unstack('device')
```

```python
#@title groupby с нормировкой по строкам (ответ)
(df.groupby(['sex', 'device'])['user_id'].count() / df.groupby('sex')['user_id'].count()).unstack('device')
```

#### 2. Кто чаще подтверждает платеж и совершает окончательную покупку?

```python
#@title crosstab (ответ)
pd.crosstab(index=df['sex'], columns=df['payment_confirmation'], normalize='index')
```

```python
#@title pivot_table (ответ)
# Используем метод pivot_table (по умолчанию рассчитывает средние значения, поэтому aggfunc='mean' можно не указывать)
df.pivot_table(index=['sex'], values='payment_confirmation')
```
